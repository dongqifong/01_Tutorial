Q1:如何描述這兩組資料的樣態?

```python
## 輸入資料
boys=[164, 176, 169, 169, 165, 175, 159, 151, 144, 160, 183, 165, 156, 170,
 164, 173, 165, 163, 177, 171]

girls=[169, 183, 170, 168, 182, 170, 173, 185, 151, 156, 162, 169, 162, 181,
 159, 154, 167, 175, 170, 160]
```

```python
import numpy as np
from scipy import stats
#Q1:如何描述這兩組資料的樣態?
# 計算統計量_平均數的方法
mean_boy=np.mean(boys)
print("男孩身高平均=",mean_boy)
mean_girl=np.mean(girls)
print("女孩身高平均=",mean_girl)

# 計算統計量_中位數的方法
np_median_boy=np.median(boys,axis=None)
print("男孩身高中位數=",np_median_boy)
np_median_girl=np.median(girls,axis=None)
print("女孩身高中位數=",np_median_girl)

#計算統計量_眾數
mode_boy=stats.mode(boys,axis=None)
print("男孩身高眾數=",mode_boy[0][0])

mode_girl=stats.mode(girls,axis=None)
print("女孩身高眾數=",mode_girl[0][0])

#計算全距:
def rangeV(x): 
  return(max(x)-min(x))
    
print("男孩身高全距=",rangeV(boys))
print("女孩身高全距=",rangeV(girls))

#計算樣本變異數
print("男孩身高變異數=",np.var(boys,ddof=1))
print("男孩身高變異數=",np.var(girls,ddof=1))


#計算樣本標準差
std_boy=np.std(boys,ddof=1)
print("男孩身高標準差=",std_boy)
std_girl=np.std(girls,ddof=1)
print("女孩身高標準差=",std_girl)

## python 百分位數
#np
print("男孩身高90百分位數=",np.percentile(boys, 90))
print("男孩身高50百分位數=",np.percentile(boys, 50))
print("男孩身高20百分位數=",np.percentile(boys, 20))

print("女孩身高90百分位數=",np.percentile(girls, 90))
print("女孩身高50百分位數=",np.percentile(girls, 50))
print("女孩身高20百分位數=",np.percentile(girls, 20))


#計算峰度和偏度
print("男孩身高偏度=",stats.skew(boys))
print("男孩身高峰度=",stats.kurtosis(boys))

print("女孩身高偏度=",stats.skew(girls))
print("女孩身高峰度=",stats.kurtosis(girls))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.hist(boys,alpha=.4,label="boy")
plt.axvline(x=mean_boy)
plt.hist(girls,color=sns.desaturate("indianred", .8),alpha=.4,label="girl")
plt.axvline(x=mean_girl,color=sns.desaturate("indianred", .8))
plt.title("all samples")
plt.legend()
plt.show()
```

# 27 用機率分布描述亂中有序的世界－ 離散型分配 (1)


```python
# 丟一個銅板，丟了100次，出現正面 50 次的機率有多大。
from scipy import stats
p = 0.5 # 假設是公平硬幣
n = 100  # 重複實驗 100次,
r = 50 # 計算出現50次正面
prob=stats.binom.pmf(r,n,p)
print(prob)
```

# 28 用機率分布描述亂中有序的世界－ 離散型分配 (2)

# 29 用機率分佈描述亂中有序的世界－連續型分配

```python
#計算標準常態分配，小於1的機率有多大?
import scipy.stats
prob=stats.norm(loc=0,scale=1).cdf(1)
print(prob)
```

```python
#計算標準常態分配，大於1，小於 -1 的機率有多大?
prob1=stats.norm(loc=0,scale=1).cdf(-1)
prob2=1-stats.norm(loc=0,scale=1).cdf(1)
prob=prob1+prob2
print(prob)
```

```python
#X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

mu=2
var=4
std=var**0.5
prob=stats.norm(loc=mu,scale=std).cdf(6)
print(prob)
```

# 33 A/B TEST

```python
#某工廠想知道兩條不同的生產線的產品不良率是否有所不同，由兩條生產線中各抽取 300 個樣本，第一組有 75 個不良品，第二組有 30 個不良品，我們可以宣稱生產線所生產出的產品不良率不相同? (以 0.05 為顯著水準)?
```

```python
import statsmodels.stats.proportion
#H0:良率相同 H1:良率不同
positive = [75,30] 
number = [300,300] 
z_value,p_value=statsmodels.stats.proportion.proportions_ztest(count=positive, nobs=number, alternative='two-sided')
print("z_value=",round(z_value,3))
print("p_vale=",round(p_value,6))
if  p_value<0.05:
    print("p_value<0.05: 故拒絕H0，良率不相同")
else:
    print("p_value>0.05: 故不拒絕H0，良率相同")
```
